```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batch_size_exp = pd.read_csv("experiments/agg_result_batch.csv")
img_size_exp = pd.read_csv("experiments/agg_result_img_size.csv")
```

```python
sec_batch = batch_size_exp.loc[batch_size_exp['plain']==0]
plain_batch = batch_size_exp.loc[batch_size_exp['plain']==1]
plt.figure(figsize=(4,2))
plt.plot(sec_batch['batch_size'],sec_batch['inference_time'],marker='v')
plt.grid(True)
plt.xlabel('Batch Size')
plt.ylabel('Inference Time (s)')
plt.savefig('../experiments/figures/sec_batch.pdf',bbox_inches='tight',pad_inches=0.1)
```

```python
plt.figure(figsize=(4,2))
plt.plot(plain_batch['batch_size'],1000*plain_batch['inference_time'],marker='o',color='#ff7f0e',linestyle='--')
plt.grid(True)
plt.xlabel('Batch Size')
plt.ylabel('Inference Time (ms)')
plt.savefig('../experiments/figures/pln_batch.pdf',bbox_inches='tight',pad_inches=0.1)
```

```python
sec_img = img_size_exp.loc[img_size_exp['plain']==0]
pln_img = img_size_exp.loc[img_size_exp['plain']==1]
plt.figure(figsize=(4,2))
plt.plot(pln_img['img_size'],1000*pln_img['inference_time'],marker='o',color='#ff7f0e',linestyle='--')
plt.grid(True)
plt.ylim(1,19)
plt.xlabel('(Square) Image Size')
plt.ylabel('Inference Time (ms)')
plt.savefig("../experiments/figures/pln_img.pdf",bbox_inches='tight',pad_inches=0.1)
```

```python
plt.figure(figsize=(4,2))
plt.plot(sec_img['img_size'],sec_img['inference_time'],marker='v')
plt.grid(True)
plt.xlabel('(Square) Image Size')
plt.ylabel('Inference Time (s)')
plt.savefig('../experiments/figures/sec_img.pdf',bbox_inches='tight',pad_inches=0.1)
```
